## Opencv with images

```python
import cv2
import numpy as np

cap= cv2.VideoCapture(0) #0 for the first webcam on your computer
#cap= cv2.VideoCapture('output.avi') -->to load external video file
#to save the captured video:
fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter('output.avi', fourcc , 20.0, (640,480))

while True: #infinite loop
    ret, frame= cap.read()  #True/False in ret, and you get a frame
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY) #convert color of frame from BGR to gray
    
    cv2.imshow('frame',frame) #named the frame as 'frame',could have given any other name
    cv2.imshow('gray',gray)
    #you need some way to get out of this infinite loop
    if cv2.waitKey(1) & 0xFF== ord('q'):
        break
cap.release()
out.release()
cv2.destroyAllWindows()    
```

## Drawing and writing on Image


```python
import numpy as np
import cv2

img= cv2.imread('watch.jpg',1)

#drawing a line on the image
cv2.line(img, (0,0), (150,150), (200,100,100), 15) #img,starting coordinates,ending coordinates,color,pixel width

#drawing a rectangle on the image
cv2.rectangle(img, (15,25), (200,150) , (0,50,0), 5)

#drawing a cirle on the image
cv2.circle(img, (100,63) , 55, (0,0,255), -1) #img,center,radius,color,width(-ve width fills in the circle)

#drawing a polygon on the image
pts = np.array([[10,5],[20,30],[70,20],[50,10]],np.int32) #points of the polygon
##pts = pts.reshape((-1,1,2))
cv2.polyLines(img,[pts], True,(0,255,255),3) #img,array of points,connect 1st and last point,color,line width

#writing on the image


cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
